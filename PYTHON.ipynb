```python
dataurl2 = 'https://www.google.com/search?ei=XcYRXIWdHYyW8gWV-LuoAQ&rlz=1C1OKWM_zh-TWTW821TW821&yv=3&tbm=isch&q={}&vet=10ahUKEwjFv7_I45vfAhUMi7wKHRX8DhUQuT0IogEoAQ.XcYRXIWdHYyW8gWV-LuoAQ.i&ved=0ahUKEwjFv7_I45vfAhUMi7wKHRX8DhUQuT0IogEoAQ&ijn=1&start={}&asearch=ichunk&async=_id:rg_s,_pms:s,_fmt:pc'
#error
dataurl = 'https://www.google.com/search?ei=StMRXOTCKoznwQPTzqzACQ&rlz=1C1OKWM_zh-TWTW821TW821&yv=3&q={}&tbm=isch&vet=10ahUKEwjkj7ry75vfAhWMc3AKHVMnC5gQuT0IogEoAQ.StMRXOTCKoznwQPTzqzACQ.i&ved=0ahUKEwjkj7ry75vfAhWMc3AKHVMnC5gQuT0IogEoAQ&ijn=1&start={}&asearch=ichunk&async=_id:rg_s,_pms:s,_fmt:pc'
def getidolImg(keyword, dstpath):
    for i in range(3):
        res= requests.get(dataurl.format(keyword, i * 100))
        soup = BeautifulSoup(res.text, 'lxml')
        for ele in soup.select('img'):
            imgurl = ele.get('src')
            fname = imgurl.split('tbn:')[1]
            with open(dstpath + fname + '.jpg', 'wb') as f:
                res2 = requests.get(imgurl)
                f.write(res2.content)
```
