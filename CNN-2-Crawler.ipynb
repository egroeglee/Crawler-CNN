```python
import os
os.mkdir('idol2_face/')
```

```python
import os
srcpath = 'idol2/'
dstpath = 'idol2_face/'
face_cascade = cv.CascadeClassifier('C:\ProgramData\Anaconda3\Library\etc\haarcascades\haarcascade_frontalface_default.xml')
for fname in os.listdir(srcpath):#迴圈
    #Print(srcpath +fname)
    img = Image.open(srcpath + fname)
    imgary = cv.imread(srcpath + fname)
    faces = face_cascade.detectMultiScale(imgary, 1.3, 5)
    if len(faces) == 1:#檢測是否只有一張臉
        x,y,w,h = faces[0]#圖片的範圍
        crpim = img.crop((x,y, x+ w, y + h )).resize((64,64)) # 修改圖片大小
        crpim.save(dstpath + fname)
    
```
